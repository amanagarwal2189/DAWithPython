Let us do a survived versus class comparison. How many survivors were there in each class

```python
df.Survived.sum()
```

```python
survivors=df[df.Survived==1]
survivors
```

```python
groupedClass=survivors.groupby('Pclass').size()
groupedClass
```

```python
groupedClass.plot(kind="bar")
```

Now let's do a class wise comparison of survivors and deads

```python
groupedClass_survivors=df.groupby(['Pclass','Survived']).size()
groupedClass_survivors
```

```python
fin=groupedClass_survivors.unstack(1)
fin
```

```python
fin.plot(kind='bar')
```

```python
fin['Total']=fin[0]+fin[1]
fin
```

```python
fin2=fin
fin2[0]=fin2[0]/fin2['Total']
fin2[1]=fin2[1]/fin2['Total']
fin2
```

```python
fin2[[0,1]].plot(kind='bar')
```

Survivor ratios based on sex and age group. Here age group is in 10s. Example 20 ->20, 25->20, 29->20, 30->30

```python
analysis2=df.groupby(['Pclass','Survived', df.Age //20 *20]).size()
analysis2.unstack(['Pclass','Survived']).fillna(0.0)
```

```python
analysis2.unstack(['Pclass','Survived']).fillna(0.0).plot(kind='bar')
```

```python
analysis2=df.groupby(['Sex','Survived', df.Age //10 *10]).size()
analysis2.unstack(['Sex','Survived']).fillna(0.0).plot(kind='bar')
```
